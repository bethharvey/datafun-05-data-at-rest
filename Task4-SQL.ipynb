```python
# Import modules
import sqlite3
import pandas as pd
```

```python
# Create connection to database
connection = sqlite3.connect('books.db')
```

```python
# Set max columns to 10
pd.options.display.max_columns = 10
```

```python
# Select everything from authors table
pd.read_sql('SELECT * FROM authors', connection, index_col=['id'])
```

```python
# Select everything from titles table
pd.read_sql('SELECT * FROM titles', connection)
```

```python
# Select everything from isbn table
pd.read_sql('SELECT * FROM author_ISBN', connection)
```

## 1 - SELECT

```python
# Get first and last names from authors table
pd.read_sql('SELECT first, last FROM authors', connection)
```

## 2 - WHERE

```python
# Get title, edition, and copyright for all books with copyright years after 2016
pd.read_sql("""SELECT title, edition, copyright
                FROM titles
                WHERE copyright > '2016'""", connection)
```

```python
# Get all authors whose last name starts with D
pd.read_sql("""SELECT id, first, last
               FROM authors
               WHERE last LIKE 'D%'""", connection, index_col=['id'])
```

```python
# Get all authors whose first name has b as the second letter
pd.read_sql("""SELECT id, first, last
               FROM authors
               WHERE first LIKE '_b%'""", connection, index_col=['id'])
```

## 3 - ORDER BY

```python
# Sort titles in ascending order
pd.read_sql('SELECT title FROM titles ORDER BY title ASC', connection)
```

```python
# Sort author names by last name, then by first name
pd.read_sql("""SELECT id, first, last
               FROM authors
               ORDER BY last, first""", connection, index_col=['id'])
```

```python
# Sort authors by last name (descending) and by first name (ascending)
pd.read_sql("""SELECT id, first, last
               FROM authors
               ORDER BY last DESC, first ASC""",
            connection, index_col=['id'])
```

```python
# Get isbn, title, edition, and copyright from titles where title ends with 'How to Program', sort by title ascending
pd.read_sql("""SELECT isbn, title, edition, copyright
               FROM titles
               WHERE title LIKE '%How to Program'
               ORDER BY title""",
            connection)
```

## 4 - INNER JOIN

```python
# Get authors and isbns for books by each author
pd.read_sql("""SELECT first, last, isbn
               FROM authors
               INNER JOIN author_ISBN
                   ON authors.id = author_ISBN.id
               ORDER BY last, first""",
            connection)
```

## 5 - INSERT INTO

```python
# Create cursor object and insert 'Sue Red' into authors table
cursor = connection.cursor()
cursor = cursor.execute("""INSERT INTO authors (first, last)
                           VALUES ('Sue', 'Red')""")
```

```python
# View updated authors table
pd.read_sql('SELECT id, first, last FROM authors', connection, index_col=['id'])
```

## 6 - UPDATE

```python
# Update 'Sue Red' to 'Sue Black'
cursor = cursor.execute("""UPDATE authors SET last = 'Black'
                           WHERE last = 'Red' AND first = 'Sue'""")
```

```python
# Check number of records updated
cursor.rowcount
```

```python
# View authors table to confirm update
pd.read_sql('SELECT id, first, last FROM authors', connection, index_col=['id'])
```

## 7 - DELETE FROM

```python
# Remove Sue Black from authors table
cursor = cursor.execute('DELETE FROM authors WHERE id = 6')
```

```python
# Check number of records updated
cursor.rowcount
```

```python
# View updated authors table
pd.read_sql('SELECT id, first, last FROM authors', connection, index_col=['id'])
```
