## 1 - Load

```python
# Import pandas
import pandas as pd
```

```python
# Load dataset as DataFrame
avonet = pd.read_csv('AVONET Data.csv')
```

This dataset is from an open-source database of physical measurements for around 11,000 species of birds.
The full dataset is called AVONET, and more information can be found in the file "AVONET Supplementary dataset1' or at https://onlinelibrary.wiley.com/doi/full/10.1111/ele.13898

## 3 - Headings

```python
# Customize column names
# Definitions can be found in the full dataset
avonet.columns = ['id', 'species', 'family', 'order', 'avibase_id', 'total_individuals','female', 'male', 'unknown', 
                  'complete_measures', 'beak_length_culmen', 'beak_length_nares', 'beak_width', 'beak_depth',
                  'tarsus_length', 'wing_length', 'kipps_distance', 'secondary', 'hand_wing_distance', 'tail_length',
                  'mass', 'mass_source', 'mass_refs_other', 'inference', 'traits_inferred', 'reference_species',
                  'habitat', 'habitat_density', 'migration', 'trophic_level', 'trophic_niche', 'primary_lifestyle',
                  'min_latitude', 'max_latitude', 'centroid_latitude', 'centroid_longitude', 'range_size']
                  
```

```python
# Change datatype of id column to int to remove '.0'
avonet.id = avonet.id.astype('Int64')

# Change datatype of id column to str to exclude from numeric analysis
avonet.id = avonet.id.astype('str')
```

```python
# View first 5 rows of updated data
avonet.head()
```
